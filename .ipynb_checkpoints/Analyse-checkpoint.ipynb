```python
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
price = pd.read_csv("price.csv", index_col=0)
water = pd.read_csv("water_level.csv", index_col=0)
water = pd.concat([water[water.columns[1:]].sum(axis=1), water['datetime']], axis=1)
water = water.rename(columns={0: "level"})
```

```python
water.datetime = pd.to_datetime(water.datetime)
price.datetime = pd.to_datetime(price.datetime)
```

```python
df = pd.merge(price, water, on='datetime', how='outer')

#Normalisation
df.level = df.level/df.level.max()
df.price = df.price/df.price.max()

df = df.sort_values('datetime')
df = df.reset_index(drop=True)
#interpolation lin√©eaire pour remplir les valeurs manquantes
df.level.interpolate(method='linear', limit_direction='both', inplace=True)
df.price.interpolate(method='linear', limit_direction='both', inplace=True)
```

```python
df.plot(x='datetime', y=['level', 'price'], figsize=(10, 5))
```

## On voit que quand le niveau des lacs/r√©servoirs diminuent, le prix de l'√©lectricit√© augmente

# On remarque que les prix n'ont pas √©t√© r√©percut√©s √† la baisse en 2017...

```python
price = price.set_index('datetime', drop=True)
water = water.set_index('datetime', drop=True)
```

```python
price_decomp = seasonal_decompose(price, model='additive',freq=12)
water_decomp = seasonal_decompose(water, model='additive',freq=365)
```

```python
price_decomp.plot()
```

# On observe dans la composante r√©siduel la mont√©e violente des prix en 2014-2016

```python
water_decomp.plot()
```
